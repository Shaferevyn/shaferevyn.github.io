## Summarizing Data

This section will summarize the data so that we can complete the tasks outlined for Project #1.

```python
# Grouping the data file by PWS Type and Contaminant

grouped_nebraska_data = nebraska_data.groupby(['pws_type' , 'contaminant']).size()

# Print the new grouped data, which is called "grouped_nebraska_data"

print(grouped_nebraska_data)
```

```python
# Creating a new data summary: Save the data to a variable by using the .reset_index(name = '')
# This code will give us the number of violations for each contaminant by the type of water system

summary = grouped_nebraska_data.reset_index(name = 'n')

# Print summary
print(summary)
```

```python
# Statistical Summary using the .agg() function. 
# This groups the data by contaminant and then finds the maximum (max), minimum (min), mean (mean), and median (median) of the data. 

stat_summary = summary.groupby('contaminant').agg(max=('n', 'max'),
                                                min=('n', 'min'),
                                                mean=('n', 'mean'),
                                                median=('n', 'median'))

print(stat_summary)
```

# Request #1

```python
#This is used to group by pws type and then find statistical data for each type
# This is used for Request 1 in the Lab Report
stat_summary = summary.groupby('pws_type').agg(max=('n', 'max'),
                                                min=('n', 'min'),
                                                mean=('n', 'mean'),
                                                median=('n', 'median'))

print(stat_summary)
```

```python
# Summary for PWS type and Contaminant using the .grouby .size() and .reset_index functions.
# This counts violations for each combination of PWS type and contaminant (this is the same summary as above).
summary = nebraska_data.groupby(['pws_type', 'contaminant']).size().reset_index(name = 'n')

# This calculates the total violations per contaminant using the .sum() function.
total_per_contaminant=summary.groupby('contaminant')['n'].sum().reset_index(name = 'total')

# This merges the total violations per contaminant back to the summary by using the .merge() function.
summary = summary.merge(total_per_contaminant, on='contaminant')

# This calculates the percentage of violations for each contaminant for each PWS type.
summary['percent_of_total'] = (summary['n'] / summary['total'])*100

print(summary)

## Creates a pivot table to summarize the data by using the function .pivot
summary_pivot = summary.pivot(index = 'contaminant', columns = 'pws_type', values = 'percent_of_total')
print(summary_pivot)

# This exports our summary_pivot to a csv
summary_pivot.to_csv('percentages.csv', index = False)
```

## Filtering Data

This section of code filters data for what we need for our analysis.

```python
#Filtering Data for only Active Systems
active_systems = nebraska_data[(nebraska_data['activity_status'] == "Active")]

print(active_systems)
```

```python
#Filtering Data for only Active System using the .query() function.

active_system_query = nebraska_data.query('activity_status == "Active"')

print(active_system_query)
```

```python
#Double Filtering: This filters the data by more than one attribute. 
#Below shows the systems which are active systems and systems with a Maximum Contaminant Level Violation.

subset_filter = nebraska_data[(nebraska_data['activity_status'] == "Active") & 
            (nebraska_data['violation_type'] == "Maximum Contaminant Level Violation, Average")]
print(subset_filter)
print(subset_filter['violation_type'])
```

```python
#Filtering the Data: This applies multiple filters
# Filtering for "Maximum Contaminant Level, Average" OR "Maximum Contaminant Level, Single Sample"
subset_filter2 = nebraska_data[(nebraska_data['activity_status']=="Active") &
                    (nebraska_data['violation_type'] == "Maximum Contaminant Level Violation, Average") |                   
                    (nebraska_data['violation_type'] == "Maximum Contaminant Level Violation,Single Sample")]

# print only the column 'violation_type
print(subset_filter2['violation_type'])

# Extracting the subset to CSV to visualize the columns
subset_filter2.to_csv('subset_filter2.csv', index=False)
```

```python
# Step 1: Filter based on Activity Status and Population 
subset_client = nebraska_data[(nebraska_data['activity_status'] == "Active") &
                (nebraska_data['population'] <= 10000)]

# Step 2: Get the violations for 1) Type of Water System 2) Source Water Type, and 3) Contaminant
subset_client_table = subset_client.groupby(['pws_type','primary_source','contaminant']).size().reset_index(name = 'n')

# Step 3: Calculate the sum of violations using the .agg() function
analysis1 = subset_client_table.groupby(['pws_type','primary_source','contaminant']).agg(sum = ('n', 'sum'))

print(analysis1)
```

# Alternative Approach Using Functions

1. Very Small: 500 or less
2. Small: 501 - 3,300
3. Medium: 3,301 - 10,000
4. Large: 10,001 - 100,000
5. Very Large: Greater than 100,000

# Request #2

```python
# This is used for Request #2 
#Define Function
# This function will define each pws by a population category
def assign_category(x):
    if x <= 500:
        return "very small"
    elif 501 <= x <= 3300:
        return "small"
    elif 3301 <= x <= 10000:
        return "medium"
    elif 10001 <= x <= 100000:
        return "large"
    else:
        return "very large"

# Apply Function
nebraska_data['pop_category'] = nebraska_data['population'].apply(assign_category)

print(nebraska_data['pop_category'])

nebraska_data.to_csv('nebraska_data_categroy.csv', index=False)
```

```python
# This code groups by pws type, contaminant, and population cateogory
# This is used for Request #2
subset2 = nebraska_data[(nebraska_data['activity_status'] == "Active")]
summary2 = nebraska_data.groupby(['pws_type', 'contaminant','pop_category']).size().reset_index(name = 'n')
summary2.to_csv("Request2_Summary.csv")
print(summary2)
```
